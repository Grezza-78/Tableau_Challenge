```python
# Dependencies and Setup
import pandas as pd
```

```python
# CitiBike data files to be combined
files_2022_2023 = ["CSV/JC-202206.csv","CSV/JC-202207.csv","CSV/JC-202208.csv","CSV/JC-202209.csv","CSV/JC-202210.csv","CSV/JC-202211.csv","CSV/JC-202212.csv","CSV/JC-202301.csv","CSV/JC-202302.csv","CSV/JC-202303.csv","CSV/JC-202304.csv","CSV/JC-202305.csv"]

# Create an empty list to store each dataframe
df_list = []

# Loop through each file in the list
for file in files_2022_2023:

    # Read the file into a dataframe
    df_22_23_years = pd.read_csv(file)

    # Append the dataframe to the list
    df_list.append(df_22_23_years)

# Concatenate all dataframes in the list into one dataframe
df_22_23_years = pd.concat(df_list)

# Show dataframe
df_22_23_years
```

```python
# Save the dataframe to a CSV file
df_22_23_years.to_csv('./Data/citibike.csv', index=False)
```
